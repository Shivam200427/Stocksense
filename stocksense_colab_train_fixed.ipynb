```python
import json, random, sys
from pathlib import Path
import joblib
import numpy as np
import pandas as pd
import tensorflow as tf
import yfinance as yf
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import Sequential
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dense, Dropout, LSTM
from tensorflow.keras.optimizers import Adam
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)
SYMBOL = 'RELIANCE.NS'; PERIOD='5y'; INTERVAL='1d'
FINAL_EPOCHS=20; FINAL_BATCH_SIZE=32; TRAIN_RATIO=0.7; VAL_RATIO=0.15
print('Python:', sys.version.split()[0])
print('TensorFlow:', tf.__version__)
print('Num GPUs Available:', len(tf.config.list_physical_devices('GPU')))
```

```python
def compute_rsi(close, period=14):
    delta = close.diff(); gain = delta.clip(lower=0); loss = -delta.clip(upper=0)
    avg_gain = gain.ewm(alpha=1/period, min_periods=period, adjust=False).mean()
    avg_loss = loss.ewm(alpha=1/period, min_periods=period, adjust=False).mean()
    rs = avg_gain / avg_loss.replace(0, np.nan)
    return 100 - (100/(1+rs))
def compute_macd(close):
    ema_fast = close.ewm(span=12, adjust=False).mean(); ema_slow = close.ewm(span=26, adjust=False).mean()
    macd = ema_fast - ema_slow; signal = macd.ewm(span=9, adjust=False).mean()
    return macd, signal, macd - signal
def fetch_and_prepare(symbol, period='5y', interval='1d'):
    df = yf.download(symbol, period=period, interval=interval, auto_adjust=False, progress=False)
    if df.empty: raise ValueError(f'No data for {symbol}')
    if isinstance(df.columns, pd.MultiIndex): df.columns = [c[0] for c in df.columns]
    df['EMA_20'] = df['Close'].ewm(span=20, adjust=False).mean()
    rm = df['Close'].rolling(20).mean(); rs = df['Close'].rolling(20).std()
    df['BB_Upper'] = rm + 2*rs; df['BB_Lower'] = rm - 2*rs
    df['RSI'] = compute_rsi(df['Close'])
    m, s, h = compute_macd(df['Close']); df['MACD'] = m; df['MACD_Signal'] = s; df['MACD_Hist'] = h
    df = df.dropna().copy()
    cols = ['Close','Volume','RSI','MACD','MACD_Signal','EMA_20','BB_Upper','BB_Lower','MACD_Hist']
    scaler = MinMaxScaler(); scaled = scaler.fit_transform(df[cols])
    return df, scaled, scaler
def build_sequences(scaled, lookback=60):
    x, y = [], []
    for i in range(lookback, len(scaled)):
        x.append(scaled[i-lookback:i]); y.append(scaled[i,0])
    return np.array(x, dtype=np.float32), np.array(y, dtype=np.float32)
def split_ts(x, y, train_ratio=0.7, val_ratio=0.15):
    n = len(x); tr = int(n*train_ratio); va = int(n*(train_ratio+val_ratio))
    return {'x_train':x[:tr],'y_train':y[:tr],'x_val':x[tr:va],'y_val':y[tr:va],'x_test':x[va:],'y_test':y[va:]}
def build_model(input_shape):
    model = Sequential([LSTM(64, return_sequences=True, input_shape=input_shape), Dropout(0.2), LSTM(32), Dropout(0.2), Dense(1)])
    model.compile(optimizer=Adam(learning_rate=1e-3), loss='mse')
    return model
```

```python
df, scaled, scaler = fetch_and_prepare(SYMBOL, PERIOD, INTERVAL)
x, y = build_sequences(scaled, lookback=60)
sp = split_ts(x, y, TRAIN_RATIO, VAL_RATIO)
model = build_model((sp['x_train'].shape[1], sp['x_train'].shape[2]))
es = EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True, verbose=0)
hist = model.fit(sp['x_train'], sp['y_train'], validation_data=(sp['x_val'], sp['y_val']), epochs=FINAL_EPOCHS, batch_size=FINAL_BATCH_SIZE, verbose=0, callbacks=[es])
pred = model.predict(sp['x_test'], verbose=0).flatten()
rmse = float(np.sqrt(mean_squared_error(sp['y_test'], pred))); mae = float(mean_absolute_error(sp['y_test'], pred))
metrics = {'val_loss': float(np.min(hist.history['val_loss'])), 'test_rmse': rmse, 'test_mae': mae}
print('METRICS:', metrics)
```

```python
import shutil
ART = BASE_DIR
model.save(ART / 'lstm_model.keras')
joblib.dump(scaler, ART / 'minmax_scaler.pkl')
with open(ART / 'metrics.json', 'w', encoding='utf-8') as f: json.dump(metrics, f, indent=2)
archive_file = shutil.make_archive('stocksense_artifacts', 'zip', root_dir=ART)
print('Created archive:', archive_file)
try:
    from google.colab import files
    files.download(archive_file)
except Exception:
    print('Not in Colab. Manually copy:', archive_file)
```
